```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import umap.umap_ as umap
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

### Carregamento do dataset de exemplo: *Iris*

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

### Redução de Dimensionalidade com PCA

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.drop('target', axis=1))
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
df['PCA1'], df['PCA2'] = pca_result[:, 0], pca_result[:, 1]

plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='target', data=df, palette='Set2')
plt.title('PCA - Componentes Principais')
plt.show()
```

### Clusterização com K-Means e Avaliação com Silhouette Score

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['kmeans_labels'] = kmeans.fit_predict(pca_result)
sil_score = silhouette_score(pca_result, df['kmeans_labels'])
print(f'Silhouette Score: {sil_score:.3f}')

plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='kmeans_labels', data=df, palette='viridis')
plt.title('K-Means Clusters sobre PCA')
plt.show()
```

### DBSCAN para Clusterização e Detecção de Outliers

```python
dbscan = DBSCAN(eps=0.7, min_samples=5)
df['dbscan_labels'] = dbscan.fit_predict(pca_result)

plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='dbscan_labels', data=df, palette='tab10')
plt.title('DBSCAN Clustering')
plt.show()
```

### Detecção de Anomalias com Isolation Forest

```python
iso = IsolationForest(contamination=0.1, random_state=42)
df['anomaly_if'] = iso.fit_predict(df_scaled)
df['anomaly_if'] = df['anomaly_if'].map({1: 'Normal', -1: 'Anomalia'})

plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='anomaly_if', data=df, palette='Set1')
plt.title('Isolation Forest - Detecção de Anomalias')
plt.show()
```

### Detecção de Anomalias com One-Class SVM

```python
ocsvm = OneClassSVM(nu=0.1, kernel='rbf', gamma='scale')
df['anomaly_svm'] = ocsvm.fit_predict(df_scaled)
df['anomaly_svm'] = df['anomaly_svm'].map({1: 'Normal', -1: 'Anomalia'})

plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='anomaly_svm', data=df, palette='coolwarm')
plt.title('One-Class SVM - Detecção de Anomalias')
plt.show()
```

### Redução com UMAP

```python
#reducer = umap.UMAP(random_state=42)
reducer = umap.UMAP()
umap_result = reducer.fit_transform(df_scaled)
df['UMAP1'], df['UMAP2'] = umap_result[:,0], umap_result[:,1]

plt.figure(figsize=(8,6))
sns.scatterplot(x='UMAP1', y='UMAP2', hue='target', data=df, palette='Set2')
plt.title('UMAP - Projeção dos Dados')
plt.show()
```
